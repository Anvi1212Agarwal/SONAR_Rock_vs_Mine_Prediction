```python
#Seperatte
X = sonar.drop(columns=60, axis=1)
Y = sonar[60]
```

```python
#Seperatte
# Print column names to check their labels
print(sonar.columns)

# Get the label of the last column
target_column_label = sonar.columns[-1]
print(f"Target column label: {target_column_label}")

# Use the correct label to drop the column and select the target
X = sonar.drop(columns=target_column_label, axis=1)
Y = sonar[target_column_label]
```

```python
print (X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
```

```python
print(X_train_predict)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
